```python
import matplotlib.pyplot as plt
sizes = [1000,2000,3000,5000,10000,25000,50000,100000]
```

```python
def func(file:str, comparisons:list, copies:list)->None:
    for line in open(file).readlines():
        a,b=line.split()
        comparisons.append(int(a))
        copies.append(int(b))
```

```python
selection_random_comparisons=[]
selection_random_copies=[]
func("selection_random_stats.txt",selection_random_comparisons,selection_random_copies)
print(len(selection_random_comparisons))
print(len(sizes))
plt.figure(figsize=(10,5))
plt.plot(sizes, selection_random_comparisons, color = "red")
plt.plot(sizes, selection_random_copies, color = "blue")
plt.legend(["Comparisons", "Copies"])
plt.title("Selection_Random")
```

```python
selection_sorted_comparisons=[]
selection_sorted_copies=[]
func("selection_sorted_stats_file.txt",selection_sorted_comparisons,selection_sorted_copies)
plt.figure(figsize=(10,5))
plt.plot(sizes, selection_sorted_comparisons, color = "red")
plt.plot(sizes, selection_sorted_copies, color = "blue")
plt.legend(["Comparisons", "Copies"])
plt.title("Selection_Sorted")
```
