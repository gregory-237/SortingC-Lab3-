```python
import matplotlib.pyplot as plt
sizes = [1000,2000,3000,5000,10000,25000,50000,100000]
```

```python
def func(file:str, comparisons:list, copies:list)->None:
    for line in open(file).readlines():
        a=line.split(' ')[0]
        b=line.split(' ')[1]
        comparisons.append(int(a))
        copies.append(int(b))
```

```python
selection_random_comparisons, selection_random_copies=[], []
selection_sorted_comparisons, selection_sorted_copies=[], []
selection_reverse_sorted_comparisons,selection_reverse_sorted_copies=[], []
func("selection_random_stats.txt",selection_random_comparisons,selection_random_copies)
func("selection_sorted_stats.txt",selection_sorted_comparisons,selection_sorted_copies)
func("selection_reverse_sorted_stats.txt",selection_reverse_sorted_comparisons,selection_reverse_sorted_copies)
fig = plt.figure(figsize=(20, 5))
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

fig.add_subplot(1, 3, 1)
plt.title('Selection Random', fontsize = 18)
plt.scatter(sizes, selection_random_comparisons, color = "red")
plt.scatter(sizes, selection_random_copies, color = "blue")
plt.legend(["Comparisons", "Copies"])
plt.xlabel('Size', fontsize = 15)
plt.ylabel('Comparisons/Copies', fontsize = 15)

fig.add_subplot(1, 3, 2)
plt.title('Selection Sorted', fontsize = 18)
plt.scatter(sizes, selection_sorted_comparisons, color = "red")
plt.scatter(sizes, selection_sorted_copies, color = "blue")
plt.legend(["Comparisons", "Copies"])
plt.xlabel('Size', fontsize = 15)
plt.ylabel('Comparisons/Copies', fontsize = 15)

fig.add_subplot(1, 3, 3)
plt.title('Selection Reverse Sorted', fontsize = 18)
plt.scatter(sizes, selection_reverse_sorted_comparisons, color = "red")
plt.scatter(sizes, selection_reverse_sorted_copies, color = "blue")
plt.legend(["Comparisons", "Copies"])
plt.xlabel('Size', fontsize = 15)
plt.ylabel('Comparisons/Copies', fontsize = 15)

plt.show()
```
